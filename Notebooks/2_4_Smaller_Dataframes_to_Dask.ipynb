```python
import pandas as pd
import dask
import dask.dataframe as dd
```

```python
ddfs = []
```

```python
for i in range(21):
    ddfs.append(dd.read_csv(f'../CSVs/word_count_csvs/wordcounts_{i}.csv',blocksize='100MB',sample=500000,
                       assume_missing=True).drop('Unnamed: 0',axis=1))
```

```python
complete_ddf = dd.concat(ddfs)
```

```python
complete_ddf.shape
```

```python
complete_ddf.to_csv('../CSVs/word_count_csvs/wordcounts_complete/')
```
